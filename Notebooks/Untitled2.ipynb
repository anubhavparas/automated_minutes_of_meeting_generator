```python
import io
import os
import subprocess
import datetime
```

```python
from google.cloud import speech
from google.cloud.speech import enums
from google.cloud.speech import types
```

```python
os.environ["GOOGLE_APPLICATION_CREDENTIALS"]="./credentials.json"
#print subprocess.Popen("sh ./setup-google-env.sh", shell=True, stdout=subprocess.PIPE).stdout.read()


#
```

```python
client = speech.SpeechClient()
audio = types.RecognitionAudio(uri='gs://mom_tip_files/wayne-olson.long.wav')
```

```python
config = types.RecognitionConfig(
        encoding=enums.RecognitionConfig.AudioEncoding.LINEAR16,
        sample_rate_hertz=16000,
        language_code='en-US',
        use_enhanced=True,
        model='phone_call'
        )
```

```python
operation = client.long_running_recognize(config, audio)

print('Waiting for op to complete...')
```

```python
response = operation.result(timeout=600)
print response
```

```python
#writing the response a file

now = str(datetime.datetime.now().strftime("%y-%m-%d-%H.%M.%S"))
file_name = "transcription_" + now + ".txt";
with open(file_name, 'w') as transcription:
    for result in response.results:
        print result.alternatives[0].transcript.replace('\\', '')
        transcription.write((result.alternatives[0].transcript.replace('\\', '') + "."))   

```
