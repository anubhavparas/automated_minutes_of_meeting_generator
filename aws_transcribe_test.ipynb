```python
from __future__ import print_function
```

```python
import time
```

```python
import boto3
```

```python
transcribe = boto3.client('transcribe')
```

```python
job_name = "transcribe-test-mp4_multispeaker1"
job_uri = "https://s3.us-east-2.amazonaws.com/ap-test-bucket-3/wayne-olson.long.wav"
job_uri2 = "https://s3.us-east-2.amazonaws.com/ap-test-bucket-3/rcastmbna200302.mp3"
job_uri3 = "https://s3.us-east-2.amazonaws.com/ap-test-bucket-3/How+Blockchain+Works+-+in+2+Minutes.mp4"
job_uri4 = "https://s3.us-east-2.amazonaws.com/ap-test-bucket-3/property-management-answering-service.mp3"
job_uri5 = "https://s3.us-east-2.amazonaws.com/ap-test-bucket-3/videoplayback.mp4"
```

```python
transcribe.start_transcription_job(
    TranscriptionJobName=job_name,
    Media={'MediaFileUri': job_uri2},
    MediaFormat='mp3',
    LanguageCode='en-US',
    Settings={
        'VocabularyName': 'default_custom_vocab',
        'ShowSpeakerLabels': True,
        'MaxSpeakerLabels': 10
    }
)
```

```python
while True:
    status = transcribe.get_transcription_job(TranscriptionJobName=job_name)
    if status['TranscriptionJob']['TranscriptionJobStatus'] in ['COMPLETED', 'FAILED']:
        break
    print("Not ready yet...")
    time.sleep(5)
print(status)
```

```python
status['TranscriptionJob']['Transcript']['TranscriptFileUri']
```

```python
import requests

transcriptionURL = status['TranscriptionJob']['Transcript']['TranscriptFileUri']
req = requests.get(transcriptionURL)

with open('transcribe-test-multi_speaker1.json', 'wb') as transcription_file:
    transcription_file.write(req.content)

```
